```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
def read_txt(filepath):
    df = pd.read_csv(filepath, sep=" ", header=None)
    df[0] = df[0].apply(lambda x: x[1:-1])
    df[0] = df[0].apply(lambda x: x.replace("\'",""))
    df[1] = df[1].apply(lambda x: int(x[0:-1]))
    print(df.head(5))
    return df
```

```python
df = pd.read_csv('./data/train.csv')
print(df.head(5))
df_dis = read_txt("./data/topWordsDisasterTweets.txt")
df_non = read_txt("./data/topWordsNonDisasterTweets.txt")
df_ign = pd.read_csv("./data/ignoreWords.txt", sep=" ", header=None)
print(df_ign.head(5))
```

```python
df_filt = ~df_dis[0].isin(df_non[0])
df_filt2 = ~df_non[0].isin(df_dis[0])
df_new = df_dis[0].loc[df_filt]
df_new_2 = df_non[0].loc[df_filt2]
print(df_new)
print(df_new_2)
print(df)
```

```python
ls_exp = []
for i in range(300):
    if i%100 ==0:
        print(i)
    exp = 0
    txt = df['text'].iloc[i]
#     print(txt)
    for j in range(1000):
        if df_new.iloc[j] in txt:
#             print(df_new.iloc[j])
            exp += 1
        if df_new_2.iloc[j] in txt:
            exp -= 1
    ls_exp.append(exp)
#     print(exp)
print(ls_exp)
```
